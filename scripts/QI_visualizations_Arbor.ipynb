Here, we are visualizing responses to the QI questions at Arbor for new patients at their first visit to the clinic. 

## Setting up the code 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from utils.file_ops import *
```

Set the clinic to be visualized here

```python
CLINIC = 'arbor'
```


Read in the results of dataprocessing

```python
results_dict = read_from_file('qi_results_dict_mom')
```

Set up the visualization environment 

```python
sns.set()
sns.set_style("white")
sns.set_context("notebook")
```

## Begin visualizations

```python
one_clinic_results = results_dict[CLINIC]
```

### First, diabetes data 

```python
one_clinic_results['QIPHQ2'].head(10)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.lineplot(data=one_clinic_results['QIDiabetes'], ax=ax, markers=True, sort=False)
plt.ylabel('Number of Patients')
plt.title('Responses to QI Diabetes at ' + CLINIC)
```

In addition to the distribution of responses indicating whether patients have diabetes, we want to look at associated questions that we would ask patients who are positive for diabetes. 

This cell summarizes labs/exams/referrals utilized for diabetic patients.

```python
db_df = read_from_file('diabetes_fused')
db_df_spec = db_df[CLINIC]
db_df_spec.head(10)
```

This is a visualization of the labs/exams/referrals utilized for diabetic patients

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.lineplot(data=db_df_spec, markers=True, dashes=False, ax=ax, sort=False)
plt.ylabel('Patient Number')
plt.title('Diabetic Patient Quality Checks at ' + CLINIC)
```

Here, we will visualize each of the individual labs/exams/referrals evaluated in QI questions against the positive diabetic patient population
